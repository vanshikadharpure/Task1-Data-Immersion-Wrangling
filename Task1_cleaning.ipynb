### Data Cleaning & Transformation :

```python
#Standardize Date Format
df['Age'] = df['Age'].fillna(df['Age'].mean())
print("Missing Age values fixed!")
print("Missing Age now:", df['Age'].isnull().sum())
```

```python
df['City'] = df['City'].fillna('Unknown')
print("Missing City values fixed!")
print("Missing City now:", df['City'].isnull().sum())
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
print("Date format fixed!")
print("Order_Date type now:", df['Order_Date'].dtype)
```

```python
#Categorizing City into Regions
def categorize_city(city):
    if city in ['Bengaluru', 'Hyderabad']:
        return 'South India'
    elif city in ['Mumbai', 'Pune']:
        return 'West India'
    elif city in ['Delhi']:
        return 'North India'
    elif city in ['Kolkata', 'Patna', 'Gaya']:
        return 'East India'
    else:
        return 'Unknown'

df['Region'] = df['City'].apply(categorize_city)
print("Region column created!")
print(df['Region'].value_counts())
```

```python
#feature engineering
def categorize_age(age):
    if age >= 18 and age <= 30:
        return 'Young'
    elif age >= 31 and age <= 50:
        return 'Middle'
    else:
        return 'Senior'

df['Age_Group'] = df['Age'].apply(categorize_age)
print("Age_Group column created!")
print(df['Age_Group'].value_counts())
print(df[['Age', 'Age_Group']].head())
```

```python
df['Order_Month'] = df['Order_Date'].dt.month
df['Order_Year'] = df['Order_Date'].dt.year
df['Order_Day'] = df['Order_Date'].dt.day

print("New columns created!")
print(df[['Order_Date', 'Order_Month', 'Order_Year', 'Order_Day']].head())
```
